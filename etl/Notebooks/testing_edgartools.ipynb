```python
import edgar
from edgar.xbrl import XBRLS
import os
from dotenv import load_dotenv

load_dotenv()
email = os.getenv("EMAIL")
```

```python
edgar.set_identity(email)
```

```python
c = edgar.Company("AAPL")

# Get data
c_filings = c.get_filings(form = "10-K", is_xbrl = True)

financials = edgar.MultiFinancials.extract(c_filings)
```

```python
income_statement = financials.income_statement()
balance_sheet = financials.balance_sheet()
cash_flows = financials.cashflow_statement()

incs = income_statement.to_dataframe()
bs = balance_sheet.to_dataframe()
cf = cash_flows.to_dataframe()
```

```python
for filing in c_filings:
    print(filing.filing_date)
```
